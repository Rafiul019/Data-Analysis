```python
##Measure of central tendency##
import numpy as np
from scipy import stats

data =[10, 12, 23, 23, 16, 23, 21, 16, 14, 13]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")

```

```python
##Measure of dispersion##
import numpy as np

data = [10, 12, 23, 23, 16, 23, 21, 16, 14, 13]

range_value = np.max(data) - np.min(data)
variance = np.var(data)
std_dev = np.std(data)

q1 = np.percentile(data,25)
q3 = np.percentile(data,75)
iqr = q3 - q1

print(f"Range: {range_value}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Interquartile Range (IQR): {iqr}")
```

```python
##Binomial Distribution##
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10  # number of trials
p = 0.5  # probability of success in each trial

# Generate binomial distribution data
x = np.arange(0, n + 1)
binomial_pmf = binom.pmf(x, n, p)

# Plot the binomial distribution
plt.figure(figsize=(10, 6))
plt.bar(x, binomial_pmf, color='skyblue', alpha=0.7)
plt.plot(x, binomial_pmf, 'o', color='darkblue')  # Add data points
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
##Poisson Distribution##
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

mu = 3  # The average number of events (λ)

# Generate Poisson distribution data
x = np.arange(0, 15)  # Range of values (0 to 14)
poisson_pmf = poisson.pmf(x, mu)

# Plot the Poisson distribution
plt.figure(figsize=(10, 6))
plt.bar(x, poisson_pmf, color='skyblue', alpha=0.7, label='Poisson PMF')
plt.plot(x, poisson_pmf, 'o', color='darkblue')  # Add data points
plt.title(f'Poisson Distribution (μ={mu})')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
mu = 3  # The average number of events (λ)

# Values for which to calculate the Poisson distribution
x = np.arange(0, 10)  # 0 through 9

# Calculate Poisson distribution (PMF) for each value in x
poisson_pmf = poisson.pmf(x, mu)

# Sum all the probabilities
total_sum = np.sum(poisson_pmf)

# Display the results
for i, p in enumerate(poisson_pmf):
    print(f"P(X={i}) = {p:.4f}")

print(f"\nSum of Poisson probabilities from 0 to 10: {total_sum:.4f}")

# Plot the Poisson distribution
plt.figure(figsize=(10, 6))
plt.bar(x, poisson_pmf, color='skyblue', alpha=0.7, label='Poisson PMF')
plt.plot(x, poisson_pmf, 'o', color='darkblue')  # Add data points
plt.title(f'Poisson Distribution (μ={mu}) for values 0 to 10')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
##Given an average of 2 soothing stars per century, what is the probability of observing exactly 4 soothing stars in a 50-year period?##
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Given data
average_per_century = 2  # Average number of soothing stars per 100 years
time_period = 50  # Time period in years
mu = (average_per_century / 100) * time_period  # Calculate λ for the 50-year period

# Probability of exactly 4 soothing stars in the 50-year period
k = 4  # Number of events
probability = poisson.pmf(k, mu)

# Display the result
print(f"Probability of observing exactly {k} soothing stars in {time_period} years: {probability:.4f}")

# Visualization: Poisson distribution for values from 0 to 10
x = np.arange(0, 11)
poisson_pmf = poisson.pmf(x, mu)

plt.figure(figsize=(10, 6))
plt.bar(x, poisson_pmf, color='skyblue', alpha=0.7, label='Poisson PMF')
plt.plot(x, poisson_pmf, 'o', color='darkblue')
plt.title(f'Poisson Distribution (μ={mu:.1f}) for Number of Soothing Stars in {time_period} Years')
plt.xlabel('Number of Soothing Stars')
plt.ylabel('Probability')
plt.grid(True)

# Highlight the probability of exactly 4 events
plt.bar(k, poisson_pmf[k], color='orange', alpha=0.7, label=f'P(X={k}) = {probability:.4f}')
plt.legend()

plt.show()
```

```python
##A space agency has a history of 80% successful launches. If they plan to launch 10 rockets, what is the probability of at least 9 successful launches?##
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Given data
n = 10  # Number of trials (rockets)
p = 0.8  # Probability of success for each launch

# Calculate the probability of exactly 9 and exactly 10 successful launches
prob_9_successes = binom.pmf(9, n, p)
prob_10_successes = binom.pmf(10, n, p)

# Calculate the probability of at least 9 successful launches
prob_at_least_9 = prob_9_successes + prob_10_successes

# Display the result
print(f"Probability of at least 9 successful launches: {prob_at_least_9:.4f}")

# Visualization: Binomial distribution for 0 to 10 successful launches
x = np.arange(0, 11)
binomial_pmf = binom.pmf(x, n, p)

plt.figure(figsize=(10, 6))
plt.bar(x, binomial_pmf, color='skyblue', alpha=0.7, label='Binomial PMF')
plt.plot(x, binomial_pmf, 'o', color='darkblue')
plt.title(f'Binomial Distribution (n={n}, p={p}) for Number of Successful Launches')
plt.xlabel('Number of Successful Launches')
plt.ylabel('Probability')
plt.grid(True)

# Highlight the probability of exactly 9 and 10 successful launches
plt.bar(9, binomial_pmf[9], color='orange', alpha=0.7, label=f'P(X=9) = {prob_9_successes:.4f}')
plt.bar(10, binomial_pmf[10], color='red', alpha=0.7, label=f'P(X=10) = {prob_10_successes:.4f}')
plt.legend()

plt.show()
```

```python
##If the average IQ is 100 with a standard deviation of 15, what is the probability of randomly selecting someone with an IQ above 130?##
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
mean = 100  # Average IQ
std_dev = 15  # Standard deviation of IQ
x = 130  # IQ value to compare

# Calculate Z-score for IQ = 130
z_score = (x - mean) / std_dev

# Calculate the probability of having an IQ above 130
prob_above_130 = 1 - norm.cdf(z_score)

# Display the result
print(f"Probability of having an IQ above {x}: {prob_above_130:.4f}")

# Visualization
# Create an array of IQ values
iq_values = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(iq_values, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(iq_values, pdf, label='Normal Distribution (mean=100, std=15)', color='blue')
plt.fill_between(iq_values[iq_values > x], pdf[iq_values > x], color='red', alpha=0.5, label=f'Area above IQ={x}')
plt.title(f'Normal Distribution of IQ with mean={mean} and std deviation={std_dev}')
plt.xlabel('IQ')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
