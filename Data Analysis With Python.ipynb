```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('/content/sample_data/train.csv')

print(train_df.head())

plt.figure(figsize=(10, 6))
sns.histplot(train_df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

gender_counts = train_df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution of Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

class_counts = train_df['Pclass'].value_counts()
plt.figure(figsize=(6, 6))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title('Class Distribution of Passengers')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

print("\nDescriptive statistics for age:")
print(train_df['Age'].describe())

print("\nGender distribution:")
print(gender_counts)

print("\nClass distribution:")
print(class_counts)
```

```python
import pandas as pd

train_df = pd.read_csv('/content/sample_data/train.csv')

number_of_survivors = train_df['Survived'].sum()
total_passengers = train_df['Survived'].count()

survival_rate = number_of_survivors / total_passengers

print(f"Number of survivors: {number_of_survivors}")
print(f"Total number of passengers: {total_passengers}")
print(f"Overall survival rate: {survival_rate:.2%}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('/content/sample_data/train.csv')

survival_by_class = train_df.groupby('Pclass')['Survived'].mean()

survival_by_gender = train_df.groupby('Sex')['Survived'].mean()

age_bins = [0, 12, 18, 35, 60, 100]
age_labels = ['Child', 'Teen', 'Adult', 'Middle Age', 'Senior']
train_df['AgeGroup'] = pd.cut(train_df['Age'], bins=age_bins, labels=age_labels)
survival_by_age_group = train_df.groupby('AgeGroup')['Survived'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_class.index, y=survival_by_class.values)
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_age_group.index, y=survival_by_age_group.values)
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()

print("Survival Rate by Class:")
print(survival_by_class)

print("\nSurvival Rate by Gender:")
print(survival_by_gender)

print("\nSurvival Rate by Age Group:")
print(survival_by_age_group)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('/content/sample_data/train.csv')

survival_by_class = train_df.groupby('Pclass')['Survived'].mean()

survival_by_gender = train_df.groupby('Sex')['Survived'].mean()

age_bins = [0, 12, 18, 35, 60, 100]
age_labels = ['Child', 'Teen', 'Adult', 'Middle Age', 'Senior']
train_df['AgeGroup'] = pd.cut(train_df['Age'], bins=age_bins, labels=age_labels)
survival_by_age_group = train_df.groupby('AgeGroup')['Survived'].mean()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_class.index, y=survival_by_class.values, palette="muted")
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_gender.index, y=survival_by_gender.values, palette="muted")
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# Plot survival rates by age group
plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_age_group.index, y=survival_by_age_group.values, palette="muted")
plt.title('Survival Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Survival Rate')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(train_df['Age'].dropna(), bins=30, kde=True, color="blue")
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

gender_counts = train_df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="muted")
plt.title('Gender Distribution of Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

class_counts = train_df['Pclass'].value_counts()
plt.figure(figsize=(6, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette="muted")
plt.title('Class Distribution of Passengers')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
